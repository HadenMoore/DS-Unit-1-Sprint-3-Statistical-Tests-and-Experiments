````
contingency = pd.crosstab(df['A16'], 
                          df['A1'])
contingency = contingency.values

print(contingency.shape)
print(contingency)
````

````
# Calculate our contingency table 
contingency_table = pd.crosstab(
    df['A1'], 
    df['A16'], 
    margins =False)

contingency_table
````

````
df['A16'].value_counts()
````

````
df['A1'].value_counts()
````

````
# Get Row Sums
row_sums = contingency_table.iloc[0:2].values
col_sums = contingency_table.iloc[0:6].values
print(row_sums)
````

````
total = 678
total
````

````
expected = []
for i in range(len(row_sums)):
  expected_row = []
  for column in col_sums:
    expected_val = column*row_sums[i]/total
    expected_row.append(expected_val)
  expected.append(expected_row)
  

expected = np.array(expected)
print(expected.shape)  
print(expected)
````

````
observed = pd.crosstab(df['A1'], df['A16']).values
print(observed.shape)
observed
````

````
chi_squared = ((contingency - expected)**2/(expected)).sum()
print(f"Chi-Squared: {chi_squared}")
````
